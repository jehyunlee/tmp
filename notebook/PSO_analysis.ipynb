```python
from IPython.display import Markdown, display

def df2md(df):    
    fmt = ['---' for i in range(len(df.columns))]
    df_fmt = pd.DataFrame([fmt], columns=df.columns)
    df_formatted = pd.concat([df_fmt, df])
    display(Markdown(df_formatted.to_csv(sep="|", index=False)))
```

```python
import os
csvpath = '.\\csvs\\'

filelist = os.listdir(csvpath)
csvlist = [file for file in filelist if file.endswith('.csv')]
datemax = 45
datelist = np.arange(datemax+1)
eval_summary = pd.DataFrame()

print(datelist)

#testlist = [file for file in csvlist if file.split('_')[2] == '0']
testlist = [file for file in csvlist if (len(file.split('_')) > 2 and file.split('_')[2] == '0')]
print(testlist)
```

```python
for i in range(20):
  testlist = [file for file in csvlist if (len(file.split('_')) > 2 and file.split('_')[2] == str(i))]
  print(testlist)
```

```python
eval_summary = pd.DataFrame(columns=['date_order', 'dim_init', 'dim_final', 'iM_final', 'timestamp'])

for date in datelist:
    testlist = [file for file in csvlist if (len(file.split('_')) > 2 and file.split('_')[2] == str(date))]
    testlistlen = len(testlist)

    dimension_final = np.array([], dtype=np.int)
    indexMinutes_final = np.array([], dtype=np.int)
        
    # Analysis on each .csv file
    for i in range(testlistlen):
        testfile = testlist[i]
        repeat = int(testfile.split('_')[3][1:])
        test_df = pd.read_csv(csvpath + '/' + testfile, header=None, na_values=[' nan'])

        test_df_dimension_init = int(test_df.head(1).loc[:, 0])
        test_df_dimension_final = int(test_df.tail(1).loc[:, 0])
        dimension_final = np.append(dimension_final, test_df_dimension_final)
        if test_df.shape[0] > 1:
          test_df_indexMinutes_final = int(test_df.tail(1).loc[:, 1:].min(axis=1))
        else:
          test_df_indexMinutes_final = np.inf
        indexMinutes_final = np.append(indexMinutes_final, test_df_indexMinutes_final)
        # print('i: {}, iM_final: {}, filename: {}'.format(i, test_df_indexMinutes_final, testfile))

    try:
      indexMinutes_final_min = indexMinutes_final.min()
      #print('date: {}, iM_final_min: {}'.format(date, indexMinutes_final_min))
      dimension_final_iMmin = dimension_final[np.argmin(indexMinutes_final)]

      #print(data[data['indexMinutes'] == indexMinutes_final_min])
      indexMinutes_final_min_idx = data[data['indexMinutes'] == indexMinutes_final_min].index.tolist()[0]
      dateAndTime = data['dateAndTime'].loc[indexMinutes_final_min_idx]
      #print(indexMinutes_final_min, dimension_final_iMmin, dateAndTime)
      summary_sub = pd.DataFrame({'date_order': [date],
                                  'dim_init': [test_df_dimension_init],
                                  'dim_final': [dimension_final_iMmin],
                                  'iM_final': [indexMinutes_final_min],
                                  'timestamp': [dateAndTime.strip('\'')]})
      eval_summary = pd.concat([eval_summary, summary_sub], ignore_index=True)
    except:
      pass

print(eval_summary)
eval_summary.to_excel('eval_summary.xlsx')
```

```python
eval_summary['dim_delta'] = eval_summary['dim_final'] - eval_summary['dim_init']
dimi = pd.DataFrame(eval_summary.loc[:,'dim_init'])
dimi['label'] = 'initial'
dimf = pd.DataFrame(eval_summary.loc[:,'dim_final'])
dimf['label'] = 'final'
#print(eval_summary.describe())
sns.set_style('white')
sns.set_context('talk')
fig = sns.distplot(eval_summary['dim_delta'], rug=True)
fig.set(xlabel='Dimension Reduction', yticklabels=[])
plt.tight_layout()
fig.get_figure().savefig('dimension_reduction.png')
```

```python
iter_sum = pd.DataFrame(columns=['date_order', 'iter', 'repeat', 'iM_min'])

for date in range(13):
  print(date)
  testlist = [file for file in csvlist if (len(file.split('_')) > 2 and file.split('_')[2] == str(date))]
  testlistlen = len(testlist)
  
  iter_sum_tmp = pd.DataFrame(columns=['date_order', 'iter', 'repeat', 'iM_min'])
    
  for i in range(testlistlen): 
    testfile = testlist[i]
    
    repeat = int(testfile.split('_')[3][1:])
    
    test_df = pd.read_csv(csvpath + testfile, header=None, na_values=[' nan'])
    iter_sum_tmp['iM_min'] = test_df.iloc[:,1:].min(axis=1).astype(int)
    iter_sum_tmp['iter'] = test_df.index.astype(int)
    iter_sum_tmp['date_order'] = int(date)
    iter_sum_tmp['repeat'] = int(repeat)
    
    iter_sum = pd.concat([iter_sum, iter_sum_tmp], axis=0)
  #df2md(iter_sum_tmp.head())

  #print(iter_sum.describe())
iter_sum['date_order'] = iter_sum['date_order'].astype('int')  
iter_sum['iM_min'] = iter_sum['iM_min'].astype('int')  
iter_sum['repeat'] = iter_sum['repeat'].astype('int')  
iter_sum['iter'] = iter_sum['iter'].astype('int')  
iter_sum.info()
```

```python
iter_sum.to_csv('iter_sum.csv', index=False)
```

* Particle Swarm Convergence vs Iteration

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
from IPython.display import display, HTML
import seaborn as sns
plt.style.use('seaborn-white')
sns.set_style('white')
sns.set_context('talk')
sns.set_palette("bright")

font_time = fm.FontProperties(size = 4)

date = 0
repeat_max = iter_sum[iter_sum['date_order'] == date]['repeat'].max() + 1
print(repeat_max)

for date in range(13):
  fig  = plt.figure(figsize=(8, 6))
  ax = fig.add_subplot(1,1,1)
  for repeat in range(repeat_max):
    iter_sum_tmp = iter_sum[iter_sum['date_order']==date]
    iter_sum_tmp = iter_sum_tmp[iter_sum_tmp['repeat']==repeat] 

    ax.plot(iter_sum_tmp['iter'], iter_sum_tmp['iM_min'], linewidth=2)
  
  plt.xlabel('Iteration')
  plt.ylabel('Time Index (min.)')
  plt.tight_layout()
  plt.savefig('./PSO/iteration_d{}.png'.format(date))
  plt.close()
```
